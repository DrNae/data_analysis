```python
def pop_year(df, year):
    '''Function that calculates total population per continent for a year input is a data frame and year output is resulting dataframe'''
    assert type(year) is int, "argument is not a number" 
    assert year>1950, "year not included in data range"
    subset_data=df[(df.year == year)][["continent","pop"]]
    popdata= subset_data.groupby("continent").sum().sort("pop")
    return popdata
print pop_year(gap_data,1952)                                
```
